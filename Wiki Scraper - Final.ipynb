```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
year = "2019"
```

```python
url="https://en.wikipedia.org/wiki/" + year +"_in_the_United_States"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
texts = [] 
#Function to just get the events
def scrapper():
    
    flag = 0
    count = 0
    for s in soup.select("h2"):
        for ns in (s.fetchNextSiblings()): 

            if (ns.text == "January[edit]" or flag == 1) and ns.name != "div" :
                flag = 1
                texts.append(ns.text)

            if ns.name == "h2":
                if count > 0:
                    #flag = 1
                    return
                else:
                    count +=1 
scrapper()
```

```python
df = pd.DataFrame(data=texts)
df.columns = ["Text"]
```

```python
years = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
    
for year in years:
    year = str(year)
    url="https://en.wikipedia.org/wiki/" + year +"_in_the_United_States"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    texts = [] 
    scrapper()
    df = pd.DataFrame(data=texts)
    df.columns = ["Text"]
    #print(df)
    for i in range(0,25,2):
        df.drop([i],inplace = True)
    df = df.reset_index(drop=True)
    df = df.assign(list=df["Text"].str.split("\n")).explode("list")
    df = df.drop("Text",axis = 1)
    months = ["January", "February", "March","April","May","June","July","August","September","October","November","December"]
    df = df.reset_index(drop = True)
    for col in df.columns:
       df[col] = df[col].str.replace("-","–")
    t = ""
    for i in range(len(df['list'])) :
        if df['list'][i].split(" ")[0] in months:
            t = df['list'][i] + " –"
        else:
            df['list'][i] = t + df['list'][i]
    df[['Date', 'Event']] = df['list'].str.split('–', n=1, expand=True)
    df = df.drop("list",axis = 1)
    df = df.dropna()
    df = df.replace(regex=[r'\[.*?\]'], value='')
    df = df.reset_index(drop = True)
    pd.set_option("display.max_rows", None, "display.max_columns", None)
    df.to_csv("Events "+year+ ".csv ", encoding = 'utf-8-sig',index = False)
```
